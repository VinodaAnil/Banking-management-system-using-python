```python
def main_menu():
    while True:
        print(' Main Menu')
        print("\n1.  Add Account")
        print('\n2.  Modify Account')
        print('\n3.  Close Account')
        print('\n4.  Transactio Menu')
        print('\n5.  Close application')
        print('\n\n')
        choice = int(input('Enter your choice ...: '))
        if choice == 1:
            add_account()
        if choice == 2:
            modify_account()
        if choice == 3:
            close_account()
        if choice ==4 :
            transaction_menu()
        if choice ==5 :
            break
def clear():
    for _ in range(65):
        print()

def add_account():
    conn=pymysql.connect(host='localhost', user='root', passwd='12345', db='bank_details')
    cur=conn.cursor()
    Account=input("Enter Your account number")
    name = input('Enter Name :')
    addr = input('Enter address ')
    phone = input('Enter Phone no :')
    email = input('Enter Email :')
    aadhar = input('Enter AAdhar no :')
    actype = input('Account Type (saving/current ) :')
    status=input('Acount status: ')
    balance = input('Enter opening balance :')
    sql = 'insert into customer(account,name,address,phone,email,aadhar_no,acc_type,balance,status) values ("'+Account+'","' + name +'","'+ addr+'","'+phone+'","'+email+'","'+aadhar+'","'+actype+'","'+balance+'","'+status+'")'
    cur.execute(sql)
    conn.commit()
    conn.close()
    print('New customer added successfully')

def modify_account():
    conn=pymysql.connect(host='localhost', user='root', passwd='12345', db='bank_details')
    cur=conn.cursor()
    acno = input('Enter customer Account No :')
    print('Modify screen ')
    print('\n 1.  Customer Name')
    print('\n 2.  Customer Address')
    print('\n 3.  Customer Phone No')
    print('\n 4.  Customer Email ID')
    choice = int(input('What do you want to change ? '))
    new_data  = input('Enter New value :')
    field_name=''
    if choice == 1:
       field_name ='name'
    if choice == 2:
       field_name = 'address'
    if choice == 3:
       field_name = 'phone'
    if choice == 4:
       field_name = 'email'
    sql ='update customer set ' + field_name + '="'+ new_data +'" where account='+acno+';' 
    print(sql)
    cur.execute(sql)
    conn.commit()
    conn.close()
    print('Customer Information modified..')
    
def close_account():
    conn=pymysql.connect(host='localhost', user='root', passwd='12345', db='bank_details')
    cur=conn.cursor()
    acno = input('Enter customer Account No :')
    sql ='update customer set status="Close" where account ='+acno+';'
    print(sql)
    cur.execute(sql)
    conn.commit()
    conn.close()
    print('Account closed')
    
def transaction_menu():
    while True:
        print(' Trasaction Menu')
        print("\n1.  Deposit Amount")
        print('\n2.  WithDraw Amount')
        print('\n3.  Back to Main Menu')
        print('\n\n')
        choice = int(input('Enter your choice ...: '))
        if choice == 1:
            deposit_amount()
        if choice == 2:
            withdraw_amount()
        if choice == 3:
            break
def deposit_amount():
    conn=pymysql.connect(host='localhost', user='root', passwd='12345', db='bank_details')
    cur=conn.cursor()
    
    acno = input('Enter account No :')
    amount = input('Enter amount :')
    sql='SELECT * FROM customer where account='+acno+'' # describing programmer table 
    query=cur.execute(sql)
    result=cur.fetchone()
    #print (result)
    if result [7]== 'Active':
        sql1 ="update customer set balance = balance+"+amount + ' where account = '+acno+' and status="active";'
        cur.execute(sql1)
        conn.commit()
        conn.close()
        #print(sql1)
        #print(sql2)
        print('\n\namount deposited')

    else:
        print('\n\nClosed or Suspended Account....')
        wait= input('\n\n\n Press any key to continue....')
        conn.close()

def withdraw_amount():
    conn=pymysql.connect(host='localhost', user='root', passwd='12345', db='bank_details')
    cur=conn.cursor()
    
    acno = input('Enter account No :')
    amount = float(input('Enter amount :'))
    amou=str(amount)
    sql='SELECT * FROM customer where account='+acno+'' # describing programmer table 
    query=cur.execute(sql)
    result=cur.fetchone()
    print (result)
    
    if result [7]== 'Active':
        if result[8]>amount:
            sql1 ="update customer set balance = balance-"+amou + ' where account = '+acno+' and status="active";'
            cur.execute(sql1)
            conn.commit()
            conn.close()
            #print(sql1)
            #print(sql2)
            print('\n\namount Withdrawn')
        else:
            print('\n\nInsufficient amount')

    else:
        print('\n\nClosed or Suspended Account....')
        wait= input('\n\n\n Press any key to continue....')
        conn.close()

        
```
